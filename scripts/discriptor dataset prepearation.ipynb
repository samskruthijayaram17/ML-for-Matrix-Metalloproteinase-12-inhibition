```python
import requests

url = "https://github.com/dataprofessor/padel/raw/main/fingerprints_xml.zip"

r = requests.get(url)

with open("padel.zip", "wb") as f:
    f.write(r.content)

print("Downloaded correctly ✅")
```

```python
with open("padel.zip", "rb") as f:
    print(f.read(4))
```

```python
import zipfile

with zipfile.ZipFile("padel.zip", 'r') as zip_ref:
    zip_ref.extractall("padel")

print("Extracted successfully ✅")
```

```python
from padelpy import padeldescriptor
import pandas as pd

# Save SMILES
df3[['canonical_smiles']].to_csv("smiles.smi", index=False, header=False)

# Run PaDEL
padeldescriptor(
    mol_dir="smiles.smi",
    d_file="descriptors_output.csv",
    removesalt=True,
    standardizenitro=True,
    fingerprints=True,
    threads=4
)

# Load results
df_desc = pd.read_csv("descriptors_output.csv")
df_desc.head()
```
